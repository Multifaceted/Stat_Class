### Histograms

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size = 2000)
```

```python
data
```

```python
# from numpy to pandas
data = pd.DataFrame(data, columns = ['a', 'b'])
```

```python
for col in 'ab':
    plt.hist(data[col])
```

```python
for col in 'ab': # density plot
    plt.hist(data[col], normed = True, alpha = .5)
```

```python
# sns.kdeplot

for col in 'ab':
    sns.kdeplot(data[col], shade = True)
```

```python
for col in 'ab':
    sns.distplot(data[col])
```

```python
sns.kdeplot(data)
```

```python
# jointplot
sns.jointplot('a', 'b', data, kind = 'kde')
```
